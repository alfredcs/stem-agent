# End-to-End AI Governance Research — Full Protocol Showcase

real-world workflow: producing an AI governance research synthesis for a

## Stage 1: Research Scout (REST API)

The Scout agent performs broad literature search using the standard
`POST /api/v1/chat` endpoint with metadata for pipeline traceability.

````python
scout_prompt = (
    f"Survey the current landscape of {TOPIC}. Cover:\n"
    "1. Key regulatory requirements (risk classification, conformity assessment, transparency)\n"
    "2. Major compliance frameworks proposed by standards bodies (ISO 42001, NIST AI RMF)\n"
    "3. Industry adoption patterns and challenges reported in 2024-2025\n"
    "4. Open technical problems (bias auditing, explainability, continuous monitoring)\n\n"
    "Cite specific regulations, standards, and papers where possible."
)

print(f"[SCOUT] Researching: {TOPIC[:60]}...\n")

if server_up:
    try:
        start = time.perf_counter()
        scout_result = client.chat(
            scout_prompt,
            caller_id="gov-scout",
            metadata={"pipeline": "ai_governance", "stage": "scout"},
            timeout=CALL_TIMEOUT,
        )
        scout_time = time.perf_counter() - start

        print_response(scout_result, show_trace=True, max_content=800)
        print(f"Completed in {scout_time:.1f}s")
        scout_output = str(scout_result.get("content", ""))
    except Exception as e:
        scout_time = 0
        scout_output = ""
        print(f"Scout failed: {type(e).__name__}: {e}")
else:
    scout_output = (
        "The EU AI Act (Regulation 2024/1689) establishes a risk-based framework...\n"
        "High-risk AI systems require: conformity assessment, risk management, "
        "data governance, transparency, human oversight, accuracy/robustness...\n"
        "ISO 42001 and NIST AI RMF provide complementary compliance structures..."
    )
    scout_time = 0
    print("[Using sample data — start server for live results]")
    print(scout_output)
````

## Stage 2: Legal Analyst via A2A Protocol

The Legal Analyst receives the Scout's output via the **A2A JSON-RPC 2.0** protocol.
This demonstrates inter-agent communication — the same protocol that would be used
if these agents were running on different servers in a mesh.

```json
{
  "jsonrpc": "2.0",
  "method": "tasks/send",
  "params": { "message": { "role": "user", "parts": [...] } }
}
```

````python
legal_prompt = (
    "You are a legal analyst specializing in AI regulation. Based on this "
    "research survey, identify the top 5 compliance gaps that organizations "
    "will face when the EU AI Act enforcement begins. For each gap:\n"
    "  - Describe the requirement\n"
    "  - Explain why it's hard to comply\n"
    "  - Suggest a practical mitigation approach\n"
    "  - Rate the compliance difficulty (1-5)\n\n"
    f"Research survey:\n{scout_output[:3000]}"
)

print("[LEGAL ANALYST] Analyzing via A2A protocol...\n")

if server_up:
    start = time.perf_counter()

    # Try A2A JSON-RPC 2.0 protocol first, fall back to REST
    try:
        a2a_result = client.a2a_send("tasks/send", {
            "message": {
                "role": "user",
                "parts": [{"type": "text", "text": legal_prompt}],
            },
        })
        legal_time = time.perf_counter() - start

        if "result" in a2a_result:
            task_data = a2a_result["result"]
            artifacts = task_data.get("artifacts", [])
            legal_output = artifacts[0]["parts"][0]["text"] if artifacts else ""
            print(f"A2A Task ID: {task_data.get('id', '?')}")
            print(f"Status: {task_data.get('status', {}).get('state', '?')}")
            print(f"Time: {legal_time:.1f}s\n")
            print(legal_output[:1000])
            if len(legal_output) > 1000:
                print("...")
        else:
            print("A2A returned non-standard response, falling back to REST...")
            resp = client.chat(legal_prompt, caller_id="gov-legal", timeout=CALL_TIMEOUT)
            legal_time = time.perf_counter() - start
            legal_output = str(resp.get("content", ""))
            print_response(resp, show_trace=False, max_content=800)
    except Exception as e:
        print(f"A2A not available ({type(e).__name__}), using REST fallback...")
        try:
            resp = client.chat(legal_prompt, caller_id="gov-legal", timeout=CALL_TIMEOUT)
            legal_time = time.perf_counter() - start
            legal_output = str(resp.get("content", ""))
            print_response(resp, show_trace=False, max_content=800)
        except Exception as e2:
            legal_time = time.perf_counter() - start
            legal_output = ""
            print(f"REST fallback also failed: {type(e2).__name__}: {e2}")
else:
    legal_output = (
        "Top 5 Compliance Gaps:\n"
        "1. Algorithmic Impact Assessment (difficulty: 4/5)...\n"
        "2. Continuous Post-Market Monitoring (difficulty: 5/5)...\n"
        "3. Training Data Governance (difficulty: 4/5)...\n"
        "4. Explainability for Complex Models (difficulty: 5/5)...\n"
        "5. Cross-Border Data Flow Compliance (difficulty: 3/5)..."
    )
    legal_time = 0
    print("[Using sample data]")
    print(legal_output)
````

## Stage 3: Technical Assessor via Async Tasks

For long-running analysis, we use the **async task API**:
1. `POST /api/v1/tasks` — create the task (returns immediately)
2. `GET /api/v1/tasks/:id` — poll for completion

This is how you'd integrate STEM Agent into a batch processing pipeline
or event-driven architecture.

````python
tech_prompt = (
    "You are a technical AI architect. Based on the compliance gaps identified, "
    "design a technical compliance architecture for a mid-size company deploying "
    "high-risk AI systems under the EU AI Act. Cover:\n"
    "1. Model risk management pipeline (development → validation → monitoring)\n"
    "2. Bias detection and mitigation toolchain\n"
    "3. Explainability layer (model-agnostic + model-specific approaches)\n"
    "4. Audit trail and documentation system\n"
    "5. Continuous monitoring and drift detection\n\n"
    f"Compliance gaps to address:\n{legal_output[:2000]}"
)

print("[TECHNICAL ASSESSOR] Creating async task...\n")

if server_up:
    # Create async task
    task = client.create_task(tech_prompt)
    task_id = task.get("id") or task.get("taskId", "?")
    print(f"Task created: {task_id}")
    print(f"Initial status: {task.get('status', '?')}")
    
    # Poll for completion
    print("\nPolling for completion:")
    tech_output = ""
    for i in range(45):  # up to 90 seconds
        time.sleep(2)
        task = client.get_task(task_id)
        status = task.get("status", "unknown")
        
        if status in ("completed", "failed"):
            print(f"  Poll {i+1}: {status}")
            tech_output = str(task.get("content", task.get("result", {}).get("content", "")))
            break
        print(f"  Poll {i+1}: {status}...", end=" ", flush=True)
    
    if tech_output:
        print(f"\nResult ({len(tech_output.split())} words):\n")
        print(tech_output[:1000])
        if len(tech_output) > 1000:
            print("...")
    else:
        # If task didn't complete in time, fall back to sync
        print("\nTask still running — falling back to synchronous call...")
        resp = client.chat(tech_prompt, caller_id="gov-technical")
        tech_output = str(resp.get("content", ""))
        print_response(resp, show_trace=False, max_content=800)
else:
    tech_output = (
        "Technical Compliance Architecture:\n"
        "1. Model Risk Pipeline: MLflow + custom validation gates...\n"
        "2. Bias Toolchain: Fairlearn + AIF360 + custom metrics...\n"
        "3. Explainability: SHAP (global) + LIME (local) + attention viz...\n"
        "4. Audit System: Immutable ML metadata store + lineage tracking...\n"
        "5. Monitoring: Evidently AI + custom drift detectors + alerting..."
    )
    print("[Using sample data]")
    print(tech_output)
````

## Stage 4: Policy Writer (Synchronous with Metadata)

The Policy Writer synthesizes all previous stages into a structured
policy document. We pass **metadata** for pipeline traceability.

````python
policy_prompt = (
    "You are a policy writer preparing an AI governance whitepaper. "
    "Synthesize the research, legal analysis, and technical architecture "
    "into a structured policy recommendation document (400-500 words).\n\n"
    "Structure:\n"
    "  - Executive Summary\n"
    "  - Regulatory Landscape\n"
    "  - Compliance Architecture Recommendations\n"
    "  - Implementation Roadmap (3 phases)\n"
    "  - Resource Requirements\n\n"
    f"Technical Architecture:\n{tech_output[:2000]}\n\n"
    f"Legal Analysis:\n{legal_output[:1000]}"
)

print("[POLICY WRITER] Synthesizing whitepaper...\n")

if server_up:
    try:
        start = time.perf_counter()
        policy_result = client.chat(
            policy_prompt,
            caller_id="gov-policy",
            metadata={
                "pipeline": "ai_governance",
                "stage": "policy_writer",
                "upstream_stages": ["scout", "legal", "technical"],
            },
            timeout=CALL_TIMEOUT,
        )
        policy_time = time.perf_counter() - start

        policy_output = str(policy_result.get("content", ""))
        print_response(policy_result, show_trace=False, max_content=1500)
        print(f"Completed in {policy_time:.1f}s")
    except Exception as e:
        policy_time = 0
        policy_output = ""
        print(f"Policy writer failed: {type(e).__name__}: {e}")
else:
    policy_output = (
        "AI Governance Whitepaper: EU AI Act Compliance Strategy\n\n"
        "Executive Summary: Organizations deploying high-risk AI systems...\n"
        "Implementation Roadmap: Phase 1 (0-6mo): Assessment & gap analysis..."
    )
    policy_time = 0
    print("[Using sample data]")
    print(policy_output)
````

## Stage 5: Executive Briefing via SSE Streaming

The final synthesis uses **Server-Sent Events (SSE) streaming** to deliver
the executive briefing in real-time. This is how you'd build a live
dashboard or chatbot interface.

```
POST /api/v1/chat/stream
→ data: {"phase": "perceive", "content": "..."}
→ data: {"phase": "reason", "content": "..."}
→ data: {"phase": "respond", "content": "..."}
→ data: [DONE]
```

````python
briefing_prompt = (
    "Write a 200-word executive briefing for the C-suite summarizing "
    "our AI governance readiness and recommended next steps. "
    "Tone: direct, confident, action-oriented. No jargon.\n\n"
    f"Policy document:\n{policy_output[:2000]}"
)

print("[EXECUTIVE BRIEFING] Streaming via SSE...\n")

if server_up:
    try:
        print("  Streaming (each . = one SSE event):")
        start = time.perf_counter()
        briefing_content = print_stream_events(
            client.stream(briefing_prompt, caller_id="gov-executive")
        )
        stream_time = time.perf_counter() - start
        
        if briefing_content:
            print(f"\n  Streamed {len(briefing_content)} chars in {stream_time:.1f}s")
            print(f"\n  Content:\n{briefing_content[:1000]}")
        else:
            print("\n  (Stream completed — content may be in final event)")
    except Exception as e:
        print(f"  Streaming not available ({e}), using sync fallback...")
        resp = client.chat(briefing_prompt, caller_id="gov-executive")
        briefing_content = str(resp.get("content", ""))
        print_response(resp, show_trace=False)
else:
    briefing_content = (
        "The EU AI Act enforcement deadline is approaching. Our assessment...\n"
        "Recommended actions: 1) Complete AI inventory by Q2...\n"
        "2) Deploy monitoring infrastructure by Q3...\n"
        "3) Submit conformity documentation by Q4..."
    )
    print("[Using sample data]")
    print(briefing_content)
````

## Stage 6: Profile Analysis & Pipeline Metrics

After the pipeline completes, we inspect how each agent's **caller profile**
has adapted via EMA-based learning.

````python
callers = ["gov-scout", "gov-legal", "gov-technical", "gov-policy", "gov-executive"]
stages = ["Scout", "Legal Analyst", "Technical Assessor", "Policy Writer", "Executive Briefing"]

if server_up:
    print("CALLER PROFILE ANALYSIS\n")
    print(f"{'Agent':<22s} {'Interactions':>12s} {'Verbosity':>10s} {'Tech Depth':>12s}")
    print("-" * 60)
    
    for caller_id, stage_name in zip(callers, stages):
        try:
            profile = client.get_profile(caller_id)
            style = profile.get("style", {})
            interactions = profile.get("interactionCount", 0)
            verbosity = style.get("verbosity", "?")
            tech_depth = style.get("technicalDepth", "?")
            print(f"{stage_name:<22s} {interactions:>12d} {str(verbosity):>10s} {str(tech_depth):>12s}")
        except Exception:
            print(f"{stage_name:<22s}  {'(not established)':>34s}")
    
    print("\nProfiles evolve with each interaction via Exponential Moving Average.")
    print("Run this pipeline multiple times to see profiles diverge.")
else:
    print("Start the server and run the pipeline to see profile evolution.")
````

## Pipeline Metrics & Visualization

````python
# Collect timing and output data
if server_up:
    outputs = [scout_output, legal_output, tech_output, policy_output, briefing_content]
    times_list = []
    for name in ["scout_time", "legal_time"]:
        times_list.append(eval(name) if name in dir() else 0)
    # Approximate missing times
    times_list.extend([0, policy_time if 'policy_time' in dir() else 0, 0])
else:
    outputs = [scout_output, legal_output, tech_output, policy_output, briefing_content]
    times_list = [0] * 5

print("PIPELINE FLOW\n")
stage_names = stages[:len(outputs)]
protocols = ["REST", "A2A", "Async Task", "REST+Meta", "SSE Stream"]

for i, (name, output, protocol) in enumerate(zip(stage_names, outputs, protocols)):
    words = len(output.split()) if output else 0
    next_stage = stage_names[i + 1] if i < len(stage_names) - 1 else "[Final Output]"
    print(f"  {name:<22s} [{protocol:>12s}]  {words:>5d} words → {next_stage}")

total_words = sum(len(o.split()) for o in outputs if o)
print(f"\n  Total output: {total_words} words across {len(stages)} stages")
print(f"  Protocols used: REST, A2A (JSON-RPC), Async Tasks, SSE Streaming")
````

## Single Agent vs Multi-Agent Comparison

How does the 5-stage pipeline compare to a single agent handling everything?

````python
single_prompt = (
    f"Produce a comprehensive AI governance analysis for {TOPIC}. "
    "Cover: regulatory landscape, compliance gaps, technical architecture, "
    "policy recommendations, and an executive briefing. "
    "Be thorough but concise (800-1000 words total)."
)

if server_up:
    print("[SINGLE AGENT] Processing entire research task...\n")
    start = time.perf_counter()
    single_result = client.chat(single_prompt, caller_id="gov-single", timeout=300)
    single_time = time.perf_counter() - start
    
    single_words = len(str(single_result.get("content", "")).split())
    pipeline_words = total_words
    
    print(f"Comparison:")
    print(f"  Pipeline (5 agents):  {pipeline_words} words, 5 perspectives, 4 protocols")
    print(f"  Single agent:         {single_words} words, 1 perspective, 1 protocol")
    print(f"\n  Single agent time:    {single_time:.1f}s")
else:
    print("Start the server to run the comparison.")

print("\nTrade-offs:")
print("  Pipeline advantages:")
print("    + Multiple specialized perspectives (legal, technical, policy)")
print("    + Built-in quality control at each handoff")
print("    + Each stage independently improvable/replaceable")
print("    + Audit trail with metadata per stage")
print("    + Demonstrates all STEM Agent protocols")
print("\n  Single agent advantages:")
print("    + Faster (one round-trip)")
print("    + Cheaper (fewer API calls)")
print("    + Simpler to debug")
````

## Protocol Reference

This notebook demonstrated all major STEM Agent communication protocols:

| Protocol | Endpoint | When to Use |
|----------|----------|-------------|
| **REST** | `POST /api/v1/chat` | Simple request-response, most common |
| **A2A** | `POST /a2a` (JSON-RPC 2.0) | Agent-to-agent communication in a mesh |
| **Async Tasks** | `POST /api/v1/tasks` | Long-running work, batch processing |
| **SSE Streaming** | `POST /api/v1/chat/stream` | Real-time UX, live dashboards |
| **Agent Card** | `GET /.well-known/agent.json` | Capability discovery, service mesh |
| **Profiles** | `GET /api/v1/profile/:id` | Behavioral adaptation monitoring |
| **Behavior** | `GET /api/v1/behavior` | Runtime configuration inspection |

### Architecture Pattern: When to Use Multi-Agent Pipelines

| Scenario | Recommendation |
|----------|---------------|
| Simple Q&A | Single agent (REST) |
| Research synthesis | Multi-agent pipeline |
| Compliance review | Multi-agent with quality gate |
| Real-time chat | Single agent (SSE streaming) |
| Background batch | Async tasks with polling |
| Agent mesh | A2A protocol for discovery + delegation |
| High-stakes analysis | Multi-agent with critic (quality matters) |
